```python
import logging
from pathlib import Path

from artifactremoval.segment import batch_process_segmentation, batch_inspect_segmentation
from artifactremoval.midas import MidasProject  
```

```python
base_dir = Path.cwd().parent
interim_directory = base_dir / "data" / "interim"
project_file = base_dir / "data" / "raw" / "MIDAS_ArtifactRemoval" / "MIDAS_ArtifactRemoval_project.xml"
project = MidasProject(project_file)

log_file = interim_directory / "05_segmentation.log"
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[
        logging.FileHandler(str(log_file), mode="w"),
        logging.StreamHandler()
    ]
)
```

```python
logging.info("Starting segmentation batch process...")
batch_process_segmentation(project, interim_directory, overwrite=True)
logging.info("Stopped batch segmentation process.")
```

```python
logging.info("Starting segmentation inspection...")
batch_inspect_segmentation(project, interim_directory, overwrite=True, save_fig=True)
logging.info("Stopped batch segmentation inspection.")
```
